## Density estimation by ABC

Let us now see the same conditional probability estimated by ABC + SM. 

```python
y_tol = np.empty((0, 18)) # images that are sufficiently close to the reference image
eps = 5e-2 # tolerance value
n_y = 100
batch_size = 10000

def generator():
    while len(y_tol) < n_y:
        yield
pbar = tqdm(generator())

# we will sample from p(y) as long as it is needed to collect 'n_y' patterns
# that are close to the reference image.
for _ in pbar:
    y = torch.rand([batch_size,18])
    x_m = model(y.cuda()).cpu().detach()
    mae = (x_m.squeeze() - x_t.unsqueeze(0)).abs().mean((1,2))
    y_eps = y[mae < eps] # images that are sufficiently close
    if len(y_eps):
        y_tol = np.concatenate([y_tol, np.array(y_eps)])
    pbar.set_postfix({'num_elements': len(y_tol)}) # track the number of elements in our array
```

```python
fig, ax = plt.subplots(3,6,figsize = [10,5])

for i in range(3):
    for j in range(6):
        sns.kdeplot(y_tol.reshape(-1,3,6)[:,i,j], fill=True, bw_adjust=0.25, label = 'Surrogate model', ax = ax[i,j]);
        ax[i,j].axvline(x=y_t.reshape(3,6)[i,j], color='k', linestyle='--', label = 'true value')
        ax[i,j].set_xlim(-0.1,1.1)
        ax[i,j].set_xlabel(' ')
        ax[i,j].set_ylabel(' ')
        
fig.tight_layout()
plt.show()
```
